```python
sharpe = [None]*5000
wts = [None]*5000
exp_rets  = [None]*5000
exp_vols = [None]*5000
for i in range(5000):
    
    weights = np.random.random(4)
    norm_wts = weights/weights.sum()
    
    for stock in [apple,cisco,ibm,amazon]:
        stock['Normalized Returns']  = stock['Adj. Close']/stock['Adj. Close'][0]
    
    for stock, ratio in zip([apple,cisco,ibm,amazon],weights):
        stock['Propo Returns']  = stock['Normalized Returns']*ratio
    for stock in [apple,cisco,ibm,amazon]:
        stock['Return Val']  = stock['Propo Returns']*1000000
    return_vals= pd.concat([apple['Return Val'],ibm['Return Val'],cisco['Return Val'],amazon['Return Val']], axis = 1)

    return_vals.columns = ['Apple','IBM','CISCO','AMAZON']
    return_vals['Sum'] = np.sum(return_vals.iloc[:,:], axis = 1)
    return_vals['Daily Change'] = np.divide(return_vals['Sum']-return_vals.shift(1)['Sum'], return_vals.shift(1)['Sum'])

    
    sharp_ratio = return_vals['Daily Change'].mean()/return_vals['Daily Change'].std() 
    annualised_SR = sharp_ratio*252**0.5
    

    exp_ret = return_vals['Daily Change'].mean()
    exp_vol = return_vals['Daily Change'].std()

    sharpe[i] = annualised_SR
    wts[i] = norm_wts
    exp_rets[i] = exp_ret
    exp_vols[i] = exp_vol
```
